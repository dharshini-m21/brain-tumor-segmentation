```python
import os, random, math, cv2
from glob import glob
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
import matplotlib.pyplot as plt
from skimage import measure, morphology

# ---------------- CONFIG ----------------
DATA_ROOT = "./"   # path with subfolders images/ and masks/
IMAGE_DIR = "images"
MASK_DIR  = "masks"

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 100
LR = 1e-4
SEED = 42
AUTOTUNE = tf.data.AUTOTUNE

os.makedirs("checkpoints", exist_ok=True)
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# ---------------- METRICS ----------------
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    inter = K.sum(y_true_f * y_pred_f)
    return (2. * inter + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1.0 - dice_coef(y_true, y_pred)

def bce_dice_loss(y_true, y_pred):
    bce = tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)
    return 0.5 * bce + 0.5 * dice_loss(y_true, y_pred)

def iou_metric(y_true, y_pred, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    inter = K.sum(y_true_f * y_pred_f)
    union = K.sum(y_true_f) + K.sum(y_pred_f) - inter
    return (inter + smooth) / (union + smooth)

# ---------------- PREPROCESSING ----------------
def refine_mask(mask):
    """Remove small noise and fill holes."""
    mask = mask.astype(np.uint8)
    mask = morphology.remove_small_objects(mask > 0, min_size=50)
    mask = morphology.remove_small_holes(mask, area_threshold=100)
    return mask.astype(np.uint8)

def extract_brain_region(img, mask):
    """Use largest mask contour to crop brain area."""
    contours = measure.label(mask)
    props = measure.regionprops(contours)
    if len(props) == 0:
        return img, mask
    largest = max(props, key=lambda x: x.area)
    minr, minc, maxr, maxc = largest.bbox
    img_crop = img[minr:maxr, minc:maxc]
    mask_crop = mask[minr:maxr, minc:maxc]
    return img_crop, mask_crop

def preprocess_image_mask(img_path, mask_path):
    """Full preprocessing: read, resize, normalize, refine, extract brain, augment optional."""
    img = cv2.imread(img_path)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    # Step 1: Resize
    img = cv2.resize(img, IMG_SIZE)
    mask = cv2.resize(mask, IMG_SIZE, interpolation=cv2.INTER_NEAREST)

    # Step 2: Normalize
    img = img / 255.0

    # Step 3: Mask refinement
    mask = refine_mask(mask)

    # Step 4: Brain extraction
    img_brain, mask_brain = extract_brain_region(img, mask)

    # If brain extraction fails (empty), revert
    if img_brain.size == 0:
        img_brain, mask_brain = img, mask

    # Step 5: Resize back to network input
    img_brain = cv2.resize(img_brain, IMG_SIZE)
    mask_brain = cv2.resize(mask_brain, IMG_SIZE, interpolation=cv2.INTER_NEAREST)

    return img, mask, img_brain, mask_brain  # return before & after

def show_before_after_preprocessing(image_paths, mask_paths, n=3):
    """Visual comparison before vs after preprocessing."""
    plt.figure(figsize=(12, 8))
    for i in range(n):
        img_b, mask_b, img_a, mask_a = preprocess_image_mask(image_paths[i], mask_paths[i])
        plt.subplot(n, 4, 4*i+1); plt.imshow(img_b[...,::-1]); plt.title("Original Image"); plt.axis('off')
        plt.subplot(n, 4, 4*i+2); plt.imshow(mask_b, cmap='gray'); plt.title("Original Mask"); plt.axis('off')
        plt.subplot(n, 4, 4*i+3); plt.imshow(img_a[...,::-1]); plt.title("Preprocessed Image"); plt.axis('off')
        plt.subplot(n, 4, 4*i+4); plt.imshow(mask_a, cmap='gray'); plt.title("Preprocessed Mask"); plt.axis('off')
    plt.tight_layout()
    plt.show()

# ---------------- DATASET CREATION ----------------
def sorted_file_list(folder):
    return sorted(glob(os.path.join(folder, "*.png")))

def get_pairs(image_dir, mask_dir):
    imgs = sorted_file_list(image_dir)
    masks = sorted_file_list(mask_dir)
    assert len(imgs) == len(masks), f"Image/mask mismatch {len(imgs)} vs {len(masks)}"
    return list(zip(imgs, masks))

def split_dataset(pairs):
    random.shuffle(pairs)
    train = pairs[:2415]; val = pairs[2415:2415+305]; test = pairs[2415+305:2415+305+613]
    return train, val, test

def load_tf(img_path, mask_path):
    img_b, mask_b, img_a, mask_a = preprocess_image_mask(img_path.numpy().decode(), mask_path.numpy().decode())
    return img_a.astype(np.float32), np.expand_dims(mask_a, -1).astype(np.float32)

def tf_preprocess(img, mask):
    img, mask = tf.py_function(func=load_tf, inp=[img, mask], Tout=(tf.float32, tf.float32))
    img.set_shape([IMG_SIZE[0], IMG_SIZE[1], 3])
    mask.set_shape([IMG_SIZE[0], IMG_SIZE[1], 1])
    return img, mask

def make_dataset(pairs, training=False):
    imgs = [p[0] for p in pairs]
    masks = [p[1] for p in pairs]
    ds = tf.data.Dataset.from_tensor_slices((imgs, masks))
    ds = ds.map(tf_preprocess, num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.shuffle(512, seed=SEED)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

# ---------------- MODEL: ENHANCED U-NET_P6 ----------------
def conv_block_enh(x, filters, kernel_size=5, dropout_rate=0.0):
    x = layers.Conv2D(filters, kernel_size, padding='same', kernel_initializer='he_normal')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    x = layers.Conv2D(filters, kernel_size, padding='same', kernel_initializer='he_normal')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.1)(x)
    if dropout_rate > 0:
        x = layers.Dropout(dropout_rate)(x)
    return x

def build_unet_p6(input_shape=(224,224,3), base_filters=32, dropout_bridge=0.3):
    inputs = layers.Input(shape=input_shape)

    c1 = conv_block_enh(inputs, base_filters)
    p1 = layers.AveragePooling2D(pool_size=(2, 2))(c1)

    c2 = conv_block_enh(p1, base_filters*2)
    p2 = layers.AveragePooling2D(pool_size=(2, 2))(c2)

    c3 = conv_block_enh(p2, base_filters*4)
    p3 = layers.AveragePooling2D(pool_size=(2, 2))(c3)

    c4 = conv_block_enh(p3, base_filters*8)
    p4 = layers.AveragePooling2D(pool_size=(2, 2))(c4)

    b  = conv_block_enh(p4, base_filters*16, dropout_rate=dropout_bridge)

    u4 = layers.Conv2DTranspose(base_filters*8, 2, strides=2, padding='same')(b)
    u4 = layers.Concatenate()([u4, c4])
    c5 = conv_block_enh(u4, base_filters*8)

    u3 = layers.Conv2DTranspose(base_filters*4, 2, strides=2, padding='same')(c5)
    u3 = layers.Concatenate()([u3, c3])
    c6 = conv_block_enh(u3, base_filters*4)

    u2 = layers.Conv2DTranspose(base_filters*2, 2, strides=2, padding='same')(c6)
    u2 = layers.Concatenate()([u2, c2])
    c7 = conv_block_enh(u2, base_filters*2)

    u1 = layers.Conv2DTranspose(base_filters, 2, strides=2, padding='same')(c7)
    u1 = layers.Concatenate()([u1, c1])
    c8 = conv_block_enh(u1, base_filters)

    out = layers.Conv2D(1, 1, activation='sigmoid')(c8)
    return models.Model(inputs, out, name="U-Net_P6_Enhanced")


# ---------------- TRAINING ----------------
def compile_model(model):
    model.compile(optimizer=RMSprop(LR),
                  loss=bce_dice_loss,
                  metrics=[dice_coef, iou_metric, tf.keras.metrics.BinaryAccuracy(name='accuracy')])
    return model

def train_model(model, train_ds, val_ds):
    ckpt = ModelCheckpoint(f"checkpoints/{model.name}_best.h5", monitor='val_dice_coef', mode='max', save_best_only=True, verbose=1)
    reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=6, verbose=1)
    early = EarlyStopping(monitor='val_dice_coef', mode='max', patience=15, restore_best_weights=True, verbose=1)
    history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=[ckpt, reduce, early])
    return history

def evaluate_model(model, test_ds):
    res = model.evaluate(test_ds)
    print(f"Loss={res[0]:.4f}, Dice={res[1]:.4f}, IoU={res[2]:.4f}, Acc={res[3]:.4f}")
    return res

# ---------------- MAIN ----------------
def main():
    pairs = get_pairs(IMAGE_DIR, MASK_DIR)
    print(f"Found {len(pairs)} image–mask pairs.")
    train_pairs, val_pairs, test_pairs = split_dataset(pairs)
    print(f"Train={len(train_pairs)}, Val={len(val_pairs)}, Test={len(test_pairs)}")

    # show before vs after preprocessing on few samples
    print("\n🧩 Showing Before vs After Preprocessing Examples...")
    show_before_after_preprocessing([p[0] for p in train_pairs[:3]], [p[1] for p in train_pairs[:3]])

    # datasets
    train_ds = make_dataset(train_pairs, training=True)
    val_ds   = make_dataset(val_pairs, training=False)
    test_ds  = make_dataset(test_pairs, training=False)

    # model
    model = build_unet_p6()
    compile_model(model)
    model.summary()

    print("\n🚀 Training Enhanced U-Net_P6 Model (with Preprocessing)")
    hist = train_model(model, train_ds, val_ds)
    print("\n✅ Evaluation on Test Set:")
    results = evaluate_model(model, test_ds)

    print("\n✅ Final Dice Score:", f"{results[1]*100:.2f}%")
    print("✅ Final Accuracy:", f"{results[3]*100:.2f}%")

if __name__ == "__main__":
    main()
```
