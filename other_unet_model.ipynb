```python
# Brain Tumor Segmentation - Complete Training Code
# Multiple Models Comparison with Full Training Implementation

import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, applications
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import random
from tqdm import tqdm
import time
import pandas as pd
from datetime import datetime
from sklearn.metrics import precision_score, recall_score, f1_score

print("TensorFlow version:", tf.__version__)

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Configuration
class Config:
    IMG_HEIGHT = 224
    IMG_WIDTH = 224
    IMG_CHANNELS = 3
    BATCH_SIZE = 32
    EPOCHS = 100
    LEARNING_RATE = 0.0001
    DROPOUT_RATE = 0.3
    NUM_CLASSES = 2  # Background and Tumor
    
config = Config()

# Data Preparation and Preprocessing
class BrainTumorDataLoader:
    def __init__(self, config):
        self.config = config
        self.images = []
        self.masks = []
        
    def load_images_from_folders(self, image_folder, mask_folder):
        """Load images and masks from specified folders"""
        print("Loading images and masks...")
        
        # Get sorted list of files
        image_files = sorted([f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))])
        mask_files = sorted([f for f in os.listdir(mask_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))])
        
        # Ensure same number of images and masks
        assert len(image_files) == len(mask_files), "Number of images and masks don't match"
        
        # Load images and masks
        for img_file, mask_file in tqdm(zip(image_files, mask_files), total=len(image_files)):
            img_path = os.path.join(image_folder, img_file)
            mask_path = os.path.join(mask_folder, mask_file)
            
            # Load and preprocess image
            image = cv2.imread(img_path)
            if image is None:
                continue
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = cv2.resize(image, (self.config.IMG_WIDTH, self.config.IMG_HEIGHT))
            image = image.astype(np.float32) / 255.0  # Normalize to [0,1]
            
            # Load and preprocess mask
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                continue
            mask = cv2.resize(mask, (self.config.IMG_WIDTH, self.config.IMG_HEIGHT))
            mask = (mask > 127).astype(np.float32)  # Binarize mask
            
            self.images.append(image)
            self.masks.append(mask)
            
        self.images = np.array(self.images)
        self.masks = np.array(self.masks)
        
        print(f"Loaded {len(self.images)} images and {len(self.masks)} masks")
        print(f"Image shape: {self.images.shape}, Mask shape: {self.masks.shape}")
        
    def prepare_datasets(self, test_size=0.2, val_size=0.1):
        """Prepare train, validation, and test datasets"""
        # Split into train+val and test
        X_train_val, X_test, y_train_val, y_test = train_test_split(
            self.images, self.masks, test_size=test_size, random_state=42
        )
        
        # Split train+val into train and val
        val_ratio = val_size / (1 - test_size)
        X_train, X_val, y_train, y_val = train_test_split(
            X_train_val, y_train_val, test_size=val_ratio, random_state=42
        )
        
        # Expand mask dimensions for model input
        y_train = np.expand_dims(y_train, axis=-1)
        y_val = np.expand_dims(y_val, axis=-1)
        y_test = np.expand_dims(y_test, axis=-1)
        
        print(f"Training set: {X_train.shape[0]} images")
        print(f"Validation set: {X_val.shape[0]} images")
        print(f"Test set: {X_test.shape[0]} images")
        
        return X_train, X_val, X_test, y_train, y_val, y_test

# Custom Metrics and Loss Functions
class Metrics:
    @staticmethod
    def dice_coefficient(y_true, y_pred, smooth=1.0):
        """Calculate Dice Similarity Coefficient (robust to 3D or 4D inputs)."""
        y_true_f = tf.keras.backend.flatten(tf.cast(y_true, tf.float32))
        y_pred_f = tf.keras.backend.flatten(tf.cast(y_pred, tf.float32))
        intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
        return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)
    
    @staticmethod
    def dice_loss(y_true, y_pred):
        """Dice loss function"""
        return 1 - Metrics.dice_coefficient(y_true, y_pred)
    
    @staticmethod
    def bce_dice_loss(y_true, y_pred):
        """Combined BCE and Dice loss"""
        bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
        dice = Metrics.dice_loss(y_true, y_pred)
        return bce + dice
    
    @staticmethod
    def iou_coefficient(y_true, y_pred, smooth=1.0):
        """Calculate Intersection over Union (robust to 3D or 4D inputs)."""
        y_true_f = tf.keras.backend.flatten(tf.cast(y_true, tf.float32))
        y_pred_f = tf.keras.backend.flatten(tf.cast(y_pred, tf.float32))
        intersection = tf.reduce_sum(y_true_f * y_pred_f)
        union = tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) - intersection
        return (intersection + smooth) / (union + smooth)
    
    @staticmethod
    def global_accuracy(y_true, y_pred):
        """Global Accuracy - proportion of correctly classified pixels (robust)."""
        y_pred_binary = tf.cast(y_pred > 0.5, tf.float32)
        y_true_f = tf.keras.backend.flatten(tf.cast(y_true, tf.float32))
        y_pred_f = tf.keras.backend.flatten(y_pred_binary)
        correct = tf.cast(tf.equal(y_true_f, y_pred_f), tf.float32)
        return tf.reduce_mean(correct)
    
    @staticmethod
    def mean_accuracy(y_true, y_pred):
        """Mean Accuracy - average accuracy across classes (robust)."""
        y_pred_binary = tf.cast(y_pred > 0.5, tf.float32)
        y_true_f = tf.keras.backend.flatten(tf.cast(y_true, tf.float32))
        y_pred_f = tf.keras.backend.flatten(y_pred_binary)

        # Compute per-class accuracies
        background_mask = tf.cast(tf.equal(y_true_f, 0.0), tf.float32)
        tumor_mask = tf.cast(tf.equal(y_true_f, 1.0), tf.float32)

        eps = tf.keras.backend.epsilon()
        acc_background = tf.reduce_sum((1.0 - y_pred_f) * background_mask) / (tf.reduce_sum(background_mask) + eps)
        acc_tumor = tf.reduce_sum(y_pred_f * tumor_mask) / (tf.reduce_sum(tumor_mask) + eps)

        return (acc_background + acc_tumor) / 2.0
    
    @staticmethod
    def weighted_iou(y_true, y_pred, smooth=1e-6):
        """Weighted Intersection over Union (robust)."""
        y_pred_binary = tf.cast(y_pred > 0.5, tf.float32)
        y_true_f = tf.keras.backend.flatten(tf.cast(y_true, tf.float32))
        y_pred_f = tf.keras.backend.flatten(y_pred_binary)

        intersection = tf.reduce_sum(y_true_f * y_pred_f)
        union = tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) - intersection

        total_pixels = tf.cast(tf.size(y_true_f), tf.float32)
        tumor_weight = tf.reduce_sum(y_true_f) / (total_pixels + tf.keras.backend.epsilon())

        iou = (intersection + smooth) / (union + smooth)
        weighted_iou = iou * tumor_weight

        return weighted_iou

# Boundary F1 Score Calculation
def calculate_boundary_f1_score(y_true, y_pred, distance_threshold=2):
    """Calculate Boundary F1 Score with distance tolerance"""
    def get_boundary_pixels(mask):
        """Extract boundary pixels from mask"""
        mask = mask.astype(np.uint8)
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        boundary_mask = np.zeros_like(mask)
        cv2.drawContours(boundary_mask, contours, -1, 1, thickness=1)
        return np.where(boundary_mask == 1)
    
    y_pred_binary = (y_pred > 0.5).astype(np.uint8)
    y_true_binary = (y_true > 0.5).astype(np.uint8)
    
    # Get boundary pixels
    pred_boundary_y, pred_boundary_x = get_boundary_pixels(y_pred_binary.squeeze())
    true_boundary_y, true_boundary_x = get_boundary_pixels(y_true_binary.squeeze())
    
    if len(pred_boundary_x) == 0 or len(true_boundary_x) == 0:
        return 0.0
    
    # Calculate distances
    true_points = np.column_stack((true_boundary_y, true_boundary_x))
    pred_points = np.column_stack((pred_boundary_y, pred_boundary_x))
    
    # Calculate precision
    precision_matches = 0
    for point in pred_points:
        distances = np.sqrt(np.sum((true_points - point)**2, axis=1))
        if np.min(distances) <= distance_threshold:
            precision_matches += 1
    precision = precision_matches / len(pred_points) if len(pred_points) > 0 else 0
    
    # Calculate recall
    recall_matches = 0
    for point in true_points:
        distances = np.sqrt(np.sum((pred_points - point)**2, axis=1))
        if np.min(distances) <= distance_threshold:
            recall_matches += 1
    recall = recall_matches / len(true_points) if len(true_points) > 0 else 0
    
    # Calculate F1 score
    if precision + recall > 0:
        f1 = 2 * precision * recall / (precision + recall)
    else:
        f1 = 0.0
    
    return f1

# ===============================
# MULTIPLE MODEL ARCHITECTURES
# ===============================

# 1. Original U-Net (Baseline)
class OriginalUNet:
    def __init__(self, config):
        self.config = config
        
    def build_model(self):
        """Build original U-Net architecture"""
        inputs = layers.Input((self.config.IMG_HEIGHT, self.config.IMG_WIDTH, self.config.IMG_CHANNELS))
        
        # Encoder
        c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
        c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
        p1 = layers.MaxPooling2D((2, 2))(c1)
        
        c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
        c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
        p2 = layers.MaxPooling2D((2, 2))(c2)
        
        c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
        c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
        p3 = layers.MaxPooling2D((2, 2))(c3)
        
        # Bridge
        c4 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(p3)
        c4 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(c4)
        
        # Decoder
        u5 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c4)
        u5 = layers.concatenate([u5, c3])
        c5 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(u5)
        c5 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c5)
        
        u6 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
        u6 = layers.concatenate([u6, c2])
        c6 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u6)
        c6 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c6)
        
        u7 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
        u7 = layers.concatenate([u7, c1])
        c7 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u7)
        c7 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c7)
        
        outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c7)
        
        model = models.Model(inputs=[inputs], outputs=[outputs])
        return model

# 2. U-Net P1 Model (Paper's Configuration)
class UNetP1:
    def __init__(self, config):
        self.config = config
        
    def build_model(self):
        """Build U-Net P1 model with kernel size 5x5, 32 channels, dropout 0.3"""
        inputs = layers.Input((self.config.IMG_HEIGHT, self.config.IMG_WIDTH, self.config.IMG_CHANNELS))
        
        # Encoder (Contracting Path)
        # Block 1
        c1 = layers.Conv2D(32, (5, 5), activation='relu', padding='same')(inputs)
        c1 = layers.Conv2D(32, (5, 5), activation='relu', padding='same')(c1)
        p1 = layers.MaxPooling2D((2, 2))(c1)
        
        # Block 2
        c2 = layers.Conv2D(64, (5, 5), activation='relu', padding='same')(p1)
        c2 = layers.Conv2D(64, (5, 5), activation='relu', padding='same')(c2)
        p2 = layers.MaxPooling2D((2, 2))(c2)
        
        # Block 3
        c3 = layers.Conv2D(128, (5, 5), activation='relu', padding='same')(p2)
        c3 = layers.Conv2D(128, (5, 5), activation='relu', padding='same')(c3)
        p3 = layers.MaxPooling2D((2, 2))(c3)
        
        # Bridge
        bridge = layers.Conv2D(256, (5, 5), activation='relu', padding='same')(p3)
        bridge = layers.Conv2D(256, (5, 5), activation='relu', padding='same')(bridge)
        bridge = layers.Dropout(0.3)(bridge)
        
        # Decoder (Expansive Path)
        # Block 4
        u4 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(bridge)
        u4 = layers.concatenate([u4, c3])
        c4 = layers.Conv2D(128, (5, 5), activation='relu', padding='same')(u4)
        c4 = layers.Conv2D(128, (5, 5), activation='relu', padding='same')(c4)
        
        # Block 5
        u5 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c4)
        u5 = layers.concatenate([u5, c2])
        c5 = layers.Conv2D(64, (5, 5), activation='relu', padding='same')(u5)
        c5 = layers.Conv2D(64, (5, 5), activation='relu', padding='same')(c5)
        
        # Block 6
        u6 = layers.Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c5)
        u6 = layers.concatenate([u6, c1])
        c6 = layers.Conv2D(32, (5, 5), activation='relu', padding='same')(u6)
        c6 = layers.Conv2D(32, (5, 5), activation='relu', padding='same')(c6)
        
        # Output layer
        outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c6)
        
        model = models.Model(inputs=[inputs], outputs=[outputs])
        return model

# 3. MobileNet Model
class MobileNetModel:
    def __init__(self, config):
        self.config = config
        
    def build_model(self):
        """Build MobileNet-based segmentation model"""
        inputs = layers.Input((self.config.IMG_HEIGHT, self.config.IMG_WIDTH, self.config.IMG_CHANNELS))
        
        # Load MobileNetV2 backbone
        base_model = applications.MobileNetV2(
            weights='imagenet',
            include_top=False,
            input_tensor=inputs
        )
        
        # Freeze base model layers
        base_model.trainable = False
        
        # Get feature maps from different levels
        skip_connection_names = [
            'block_1_expand_relu',   # 112x112
            'block_3_expand_relu',   # 56x56
            'block_6_expand_relu',   # 28x28
            'block_13_expand_relu',  # 14x14
        ]
        
        # Get skip connections
        skip_connections = [base_model.get_layer(name).output for name in skip_connection_names]
        model_output = base_model.output
        
        # Decoder
        x = layers.Conv2D(256, 1, padding='same')(model_output)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        
        # Upsampling with skip connections
        for i, skip in enumerate(reversed(skip_connections)):
            x = layers.UpSampling2D((2, 2))(x)
            x = layers.Conv2D(256 // (2 ** i), 3, padding='same')(x)
            x = layers.BatchNormalization()(x)
            x = layers.Activation('relu')(x)
            
            # Add skip connection
            skip = layers.Conv2D(256 // (2 ** i), 1, padding='same')(skip)
            x = layers.concatenate([x, skip])
            
            x = layers.Conv2D(256 // (2 ** i), 3, padding='same')(x)
            x = layers.BatchNormalization()(x)
            x = layers.Activation('relu')(x)
        
        # Final upsampling to original size
        x = layers.UpSampling2D((2, 2))(x)
        x = layers.Conv2D(32, 3, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        
        outputs = layers.Conv2D(1, 1, activation='sigmoid')(x)
        
        model = models.Model(inputs=inputs, outputs=outputs)
        return model

# 4. DeepLabV3 Models
class DeepLabV3Models:
    def __init__(self, config):
        self.config = config
        
    def build_deeplabv3_resnet50(self):
        """Build simplified DeepLabV3 with ResNet50 backbone (fixed upsampling)"""
        inputs = layers.Input((self.config.IMG_HEIGHT, self.config.IMG_WIDTH, self.config.IMG_CHANNELS))
        
        # Load ResNet50 backbone
        base_model = applications.ResNet50(
            weights='imagenet',
            include_top=False,
            input_tensor=inputs
        )
        
        # Get the output of ResNet50
        x = base_model.output  # shape: (batch, H/32, W/32, channels) for typical ResNet50
        
        # Simple ASPP (a lightweight version)
        b1 = layers.Conv2D(256, 1, padding='same', activation='relu')(x)
        b2 = layers.Conv2D(256, 3, padding='same', dilation_rate=6, activation='relu')(x)
        b3 = layers.Conv2D(256, 3, padding='same', dilation_rate=12, activation='relu')(x)
        b4 = layers.Conv2D(256, 3, padding='same', dilation_rate=18, activation='relu')(x)
        
        # Concatenate and process
        x = layers.concatenate([b1, b2, b3, b4])
        x = layers.Conv2D(256, 1, activation='relu', padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        
        # Upsample back to original image size.
        # For ResNet50 with input size 224x224, final feature map is typically 7x7 (224/32 = 7).
        # So we upsample by factor 32 to return to 224x224.
        upsample_factor = (self.config.IMG_HEIGHT // tf.keras.backend.int_shape(x)[1],
                           self.config.IMG_WIDTH // tf.keras.backend.int_shape(x)[2])
        # safe fallback if shape inference fails:
        if upsample_factor[0] is None or upsample_factor[1] is None:
            upsample_factor = (32, 32)
        
        x = layers.UpSampling2D(size=upsample_factor, interpolation='bilinear')(x)
        x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        
        outputs = layers.Conv2D(1, 1, activation='sigmoid', padding='same')(x)
        
        model = models.Model(inputs=inputs, outputs=outputs)
        return model

# Model Training and Evaluation Class
class BrainTumorSegmenter:
    def __init__(self, config):
        self.config = config
        self.models = {}
        self.histories = {}
        self.results = {}
        self.training_times = {}
        
    def build_all_models(self):
        """Build all models for comparison"""
        print("Building all models...")
        
        # 1. Original U-Net
        original_unet = OriginalUNet(self.config)
        self.models['Original_U-Net'] = original_unet.build_model()
        
        # 2. U-Net P1 Model
        unet_p1 = UNetP1(self.config)
        self.models['U-Net_P1'] = unet_p1.build_model()
        
        # 3. MobileNet Model
        mobilenet_model = MobileNetModel(self.config)
        self.models['MobileNet'] = mobilenet_model.build_model()
        
        # 4. DeepLabV3 Models
        deeplab_models = DeepLabV3Models(self.config)
        self.models['DeepLabV3_ResNet50'] = deeplab_models.build_deeplabv3_resnet50()
        
        # Compile all models
        for name, model in self.models.items():
            model.compile(
                optimizer=tf.keras.optimizers.Adam(learning_rate=self.config.LEARNING_RATE),
                loss=Metrics.bce_dice_loss,
                metrics=['accuracy', Metrics.dice_coefficient, Metrics.iou_coefficient,
                        Metrics.global_accuracy, Metrics.mean_accuracy, Metrics.weighted_iou]
            )
            print(f"✅ {name} built successfully! Parameters: {model.count_params():,}")
        
        return self.models
    
    def calculate_all_metrics(self, y_true, y_pred):
        """Calculate all evaluation metrics mentioned in the paper"""
        y_true_flat = y_true.flatten()
        y_pred_binary = (y_pred > 0.5).astype(np.float32)
        y_pred_flat = y_pred_binary.flatten()
        
        # Basic metrics: use TensorFlow-based functions but pass single-sample arrays
        dice = Metrics.dice_coefficient(
            tf.convert_to_tensor(y_true, dtype=tf.float32),
            tf.convert_to_tensor(y_pred, dtype=tf.float32)
        ).numpy()
        
        iou = Metrics.iou_coefficient(
            tf.convert_to_tensor(y_true, dtype=tf.float32),
            tf.convert_to_tensor(y_pred, dtype=tf.float32)
        ).numpy()
        
        global_acc = Metrics.global_accuracy(
            tf.convert_to_tensor(y_true, dtype=tf.float32),
            tf.convert_to_tensor(y_pred, dtype=tf.float32)
        ).numpy()
        
        mean_acc = Metrics.mean_accuracy(
            tf.convert_to_tensor(y_true, dtype=tf.float32),
            tf.convert_to_tensor(y_pred, dtype=tf.float32)
        ).numpy()
        
        weighted_iou = Metrics.weighted_iou(
            tf.convert_to_tensor(y_true, dtype=tf.float32),
            tf.convert_to_tensor(y_pred, dtype=tf.float32)
        ).numpy()
        
        # Boundary F1 Score
        boundary_f1 = calculate_boundary_f1_score(y_true, y_pred)
        
        # Additional metrics (sklearn)
        precision = precision_score(y_true_flat, y_pred_flat, zero_division=0)
        recall = recall_score(y_true_flat, y_pred_flat, zero_division=0)
        specificity = recall_score(1 - y_true_flat, 1 - y_pred_flat, zero_division=0)
        
        return {
            'Global_Accuracy': global_acc,
            'Mean_Accuracy': mean_acc,
            'Mean_IoU': iou,
            'Weighted_IoU': weighted_iou,
            'Mean_Boundary_F1': boundary_f1,
            'Dice_Coefficient': dice,
            'Precision': precision,
            'Recall': recall,
            'Specificity': specificity
        }
    
    def train_models(self, X_train, y_train, X_val, y_val, epochs_per_model=50):
        """Train all models with progress tracking"""
        print(f"\nTraining all models for {epochs_per_model} epochs each...")
        
        for name, model in self.models.items():
            print(f"\n{'='*60}")
            print(f"🚀 Training {name}...")
            print(f"{'='*60}")
            
            start_time = time.time()
            
            # Enhanced callbacks for each model
            callbacks = [
                tf.keras.callbacks.EarlyStopping(
                    monitor='val_dice_coefficient',
                    patience=15,
                    restore_best_weights=True,
                    mode='max',
                    verbose=1
                ),
                tf.keras.callbacks.ReduceLROnPlateau(
                    monitor='val_dice_coefficient',
                    factor=0.5,
                    patience=8,
                    min_lr=1e-7,
                    mode='max',
                    verbose=1
                ),
                tf.keras.callbacks.ModelCheckpoint(
                    f'best_{name.replace(" ", "_").lower()}.h5',
                    monitor='val_dice_coefficient',
                    save_best_only=True,
                    mode='max',
                    verbose=1
                )
            ]
            
            # Data augmentation for training
            datagen = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.1,
                height_shift_range=0.1,
                horizontal_flip=True,
                vertical_flip=True,
                fill_mode='constant'
            )
            
            # Train model with augmentation
            history = model.fit(
                datagen.flow(X_train, y_train, batch_size=self.config.BATCH_SIZE),
                steps_per_epoch=max(1, len(X_train) // self.config.BATCH_SIZE),
                epochs=epochs_per_model,
                validation_data=(X_val, y_val),
                callbacks=callbacks,
                verbose=1
            )
            
            training_time = time.time() - start_time
            self.histories[name] = history
            self.training_times[name] = training_time
            
            print(f"✅ {name} training completed in {training_time/60:.2f} minutes")
            
            # Plot training history for this model
            self.plot_single_model_history(name, history)
        
        return self.histories
    
    def evaluate_models(self, X_test, y_test):
        """Evaluate all models on test set with all metrics"""
        print(f"\n{'='*60}")
        print("📊 Evaluating all models on test set...")
        print(f"{'='*60}")
        
        self.results = {}
        
        for name, model in self.models.items():
            print(f"\nEvaluating {name}...")
            try:
                # Get predictions
                y_pred = model.predict(X_test, verbose=0)
                
                # Calculate all metrics
                all_metrics = {}
                boundary_f1_scores = []
                
                # Calculate metrics for each sample and average
                for i in range(len(X_test)):
                    sample_metrics = self.calculate_all_metrics(y_test[i], y_pred[i])
                    boundary_f1_scores.append(sample_metrics['Mean_Boundary_F1'])
                    
                    if i == 0:  # Initialize with first sample
                        for key in sample_metrics.keys():
                            if key != 'Mean_Boundary_F1':
                                all_metrics[key] = sample_metrics[key]
                    else:  # Accumulate for averaging
                        for key in sample_metrics.keys():
                            if key != 'Mean_Boundary_F1':
                                all_metrics[key] += sample_metrics[key]
                
                # Average the metrics
                for key in all_metrics.keys():
                    all_metrics[key] /= len(X_test)
                
                # Add mean boundary F1 score
                all_metrics['Mean_Boundary_F1'] = np.mean(boundary_f1_scores)
                
                # Add training time
                all_metrics['Training_Time_Minutes'] = self.training_times.get(name, 0) / 60
                
                self.results[name] = all_metrics
                
                print(f"✅ {name} Results:")
                print(f"   📈 Global Accuracy: {all_metrics['Global_Accuracy']*100:.2f}%")
                print(f"   🎯 Mean Accuracy: {all_metrics['Mean_Accuracy']*100:.2f}%")
                print(f"   🔷 Mean IoU: {all_metrics['Mean_IoU']*100:.2f}%")
                print(f"   ⚖️  Weighted IoU: {all_metrics['Weighted_IoU']*100:.2f}%")
                print(f"   📏 Mean Boundary F1: {all_metrics['Mean_Boundary_F1']*100:.2f}%")
                print(f"   🎲 Dice Coefficient: {all_metrics['Dice_Coefficient']*100:.2f}%")
                print(f"   🎯 Precision: {all_metrics['Precision']*100:.2f}%")
                print(f"   🔄 Recall: {all_metrics['Recall']*100:.2f}%")
                print(f"   🛡️  Specificity: {all_metrics['Specificity']*100:.2f}%")
                print(f"   ⏱️  Training Time: {all_metrics['Training_Time_Minutes']:.2f} min")
                
            except Exception as e:
                print(f"❌ Error evaluating {name}: {e}")
        
        return self.results
    
    def plot_single_model_history(self, model_name, history):
        """Plot training history for a single model"""
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))
        fig.suptitle(f'Training History - {model_name}', fontsize=16, fontweight='bold')
        
        metrics_to_plot = [
            ('loss', 'Loss'),
            ('accuracy', 'Accuracy'),
            ('dice_coefficient', 'Dice Coefficient'),
            ('iou_coefficient', 'IoU Coefficient'),
            ('global_accuracy', 'Global Accuracy'),
            ('mean_accuracy', 'Mean Accuracy')
        ]
        
        for idx, (metric, title) in enumerate(metrics_to_plot):
            row = idx // 3
            col = idx % 3
            
            if metric in history.history:
                axes[row, col].plot(history.history[metric], label=f'Training {title}', linewidth=2)
                if f'val_{metric}' in history.history:
                    axes[row, col].plot(history.history[f'val_{metric}'], label=f'Validation {title}', linewidth=2)
                axes[row, col].set_title(title)
                axes[row, col].set_ylabel(title)
                axes[row, col].set_xlabel('Epoch')
                axes[row, col].legend()
                axes[row, col].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def compare_models(self):
        """Compare all models and display results"""
        if not self.results:
            print("❌ No results available. Please train and evaluate models first.")
            return
        
        print(f"\n{'='*100}")
        print("🏆 MODEL COMPARISON RESULTS")
        print(f"{'='*100}")
        
        # Sort models by Dice coefficient (main metric)
        sorted_models = sorted(self.results.items(), key=lambda x: x[1]['Dice_Coefficient'], reverse=True)
        
        headers = ['Model', 'Global Acc', 'Mean Acc', 'Mean IoU', 'Weighted IoU', 'MeanBF', 'Dice', 'Precision', 'Recall', 'Specificity', 'Time (min)']
        print(f"{headers[0]:<15} {headers[1]:<12} {headers[2]:<12} {headers[3]:<12} {headers[4]:<12} {headers[5]:<12} {headers[6]:<12} {headers[7]:<12} {headers[8]:<12} {headers[9]:<12} {headers[10]:<12}")
        print(f"{'-'*140}")
        
        for name, metrics in sorted_models:
            print(f"{name:<15} {metrics['Global_Accuracy']*100:>10.2f}% {metrics['Mean_Accuracy']*100:>10.2f}% "
                  f"{metrics['Mean_IoU']*100:>10.2f}% {metrics['Weighted_IoU']*100:>10.2f}% {metrics['Mean_Boundary_F1']*100:>10.2f}% "
                  f"{metrics['Dice_Coefficient']*100:>10.2f}% {metrics['Precision']*100:>10.2f}% {metrics['Recall']*100:>10.2f}% "
                  f"{metrics['Specificity']*100:>10.2f}% {metrics['Training_Time_Minutes']:>10.2f}")
        
        # Find best model
        best_model_name, best_metrics = sorted_models[0]
        print(f"\n🎉 BEST MODEL: {best_model_name}")
        print(f"   📊 Global Accuracy: {best_metrics['Global_Accuracy']*100:.2f}%")
        print(f"   🎯 Mean Accuracy: {best_metrics['Mean_Accuracy']*100:.2f}%")
        print(f"   🔷 Mean IoU: {best_metrics['Mean_IoU']*100:.2f}%")
        print(f"   ⚖️  Weighted IoU: {best_metrics['Weighted_IoU']*100:.2f}%")
        print(f"   📏 Mean Boundary F1: {best_metrics['Mean_Boundary_F1']*100:.2f}%")
        print(f"   🎲 Dice Coefficient: {best_metrics['Dice_Coefficient']*100:.2f}%")
        print(f"   ⏱️  Training Time: {best_metrics['Training_Time_Minutes']:.2f} minutes")
        
        return best_model_name, best_metrics
    
    def plot_comparison(self):
        """Plot comparison of all models"""
        if not self.results:
            print("❌ No results available. Please train and evaluate models first.")
            return
        
        models = list(self.results.keys())
        
        # Key metrics for comparison
        key_metrics = {
            'Global_Accuracy': 'Global Accuracy (%)',
            'Mean_Accuracy': 'Mean Accuracy (%)', 
            'Mean_IoU': 'Mean IoU (%)',
            'Weighted_IoU': 'Weighted IoU (%)',
            'Mean_Boundary_F1': 'Mean Boundary F1 (%)',
            'Dice_Coefficient': 'Dice Coefficient (%)',
            'Training_Time_Minutes': 'Training Time (minutes)'
        }
        
        fig, axes = plt.subplots(2, 4, figsize=(25, 12))
        axes = axes.flatten()
        
        for idx, (metric, title) in enumerate(key_metrics.items()):
            if idx >= len(axes):
                break
                
            values = [self.results[name][metric] * 100 if metric != 'Training_Time_Minutes' else self.results[name][metric] for name in models]
            colors = plt.cm.Set3(np.linspace(0, 1, len(models)))
            
            bars = axes[idx].bar(models, values, color=colors, alpha=0.7, edgecolor='black', linewidth=1)
            axes[idx].set_title(title, fontsize=12, fontweight='bold')
            axes[idx].set_ylabel(title)
            axes[idx].tick_params(axis='x', rotation=45)
            axes[idx].grid(True, alpha=0.3, axis='y')
            
            # Add value labels on bars
            for bar in bars:
                height = bar.get_height()
                axes[idx].text(bar.get_x() + bar.get_width()/2., height + 0.5,
                             f'{height:.1f}', ha='center', va='bottom', fontweight='bold', fontsize=8)
        
        # Hide unused subplots
        for idx in range(len(key_metrics), len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.show()
        
        # Create results dataframe
        results_df = pd.DataFrame.from_dict(self.results, orient='index')
        results_df = results_df.sort_values('Dice_Coefficient', ascending=False)
        print(f"\n📋 Detailed Results DataFrame:")
        print(results_df.round(4))
        
        return results_df

# Create demo dataset
def create_demo_dataset(num_samples=3064):
    """Create a demo dataset with synthetic brain tumor images"""
    print(f"Creating demo dataset with {num_samples} samples...")
    
    images = []
    masks = []
    
    for i in tqdm(range(num_samples), desc="Generating synthetic data"):
        # Create synthetic image (simulating MRI)
        img = np.random.rand(config.IMG_HEIGHT, config.IMG_WIDTH, 3) * 0.3
        
        # Add some brain-like structures
        center_x, center_y = config.IMG_WIDTH // 2, config.IMG_HEIGHT // 2
        y, x = np.ogrid[:config.IMG_HEIGHT, :config.IMG_WIDTH]
        
        # Create elliptical brain structure
        brain_mask = ((x - center_x)**2 / (center_x**2) + (y - center_y)**2 / (center_y**2)) <= 1
        img[brain_mask] += 0.4
        
        # Add synthetic tumor (more realistic shape)
        if np.random.random() > 0.3:  # 70% chance of having a tumor
            tumor_x = np.random.randint(center_x - 40, center_x + 40)
            tumor_y = np.random.randint(center_y - 40, center_y + 40)
            tumor_radius_x = np.random.randint(15, 35)
            tumor_radius_y = np.random.randint(15, 35)
            
            # Elliptical tumor
            tumor_mask = ((x - tumor_x)**2 / tumor_radius_x**2 + (y - tumor_y)**2 / tumor_radius_y**2) <= 1
            img[tumor_mask] += 0.3
            
            # Create corresponding mask
            mask = np.zeros((config.IMG_HEIGHT, config.IMG_WIDTH))
            mask[tumor_mask] = 1.0
        else:
            mask = np.zeros((config.IMG_HEIGHT, config.IMG_WIDTH))
        
        images.append(img.astype(np.float32))
        masks.append(mask.astype(np.float32))
    
    return np.array(images), np.array(masks)

# Visualization functions
def visualize_sample_predictions(segmenter, X_test, y_test, num_samples=3):
    """Visualize predictions from all models on sample images"""
    print(f"\n{'='*60}")
    print("👁️  Visualizing Sample Predictions")
    print(f"{'='*60}")
    
    for sample_idx in range(min(num_samples, len(X_test))):
        fig, axes = plt.subplots(3, len(segmenter.models) + 2, figsize=(20, 12))
        fig.suptitle(f'Sample {sample_idx + 1} - Model Predictions Comparison', fontsize=16, fontweight='bold')
        
        # Original image
        axes[0, 0].imshow(X_test[sample_idx])
        axes[0, 0].set_title('Original Image', fontweight='bold')
        axes[0, 0].axis('off')
        
        # Ground truth
        axes[1, 0].imshow(y_test[sample_idx].squeeze(), cmap='gray')
        axes[1, 0].set_title('Ground Truth', fontweight='bold')
        axes[1, 0].axis('off')
        
        # Empty cell for alignment
        axes[2, 0].axis('off')
        
        # Model predictions
        for idx, (model_name, model) in enumerate(segmenter.models.items(), 1):
            try:
                prediction = model.predict(X_test[sample_idx:sample_idx+1], verbose=0)[0]
                
                # Prediction mask
                axes[0, idx].imshow(prediction.squeeze(), cmap='gray')
                axes[0, idx].set_title(f'{model_name}\nPrediction', fontweight='bold')
                axes[0, idx].axis('off')
                
                # Overlay
                axes[1, idx].imshow(X_test[sample_idx])
                axes[1, idx].imshow(prediction.squeeze(), alpha=0.5, cmap='jet')
                axes[1, idx].set_title(f'{model_name}\nOverlay', fontweight='bold')
                axes[1, idx].axis('off')
                
                # Difference from ground truth
                diff = np.abs(prediction.squeeze() - y_test[sample_idx].squeeze())
                axes[2, idx].imshow(diff, cmap='hot')
                axes[2, idx].set_title(f'{model_name}\nDifference', fontweight='bold')
                axes[2, idx].axis('off')
                
            except Exception as e:
                print(f"❌ Error generating prediction for {model_name}: {e}")
                for i in range(3):
                    axes[i, idx].text(0.5, 0.5, 'Error', transform=axes[i, idx].transAxes, 
                                    ha='center', va='center', fontweight='bold')
                    axes[i, idx].axis('off')
        
        plt.tight_layout()
        plt.show()

# Main execution function
def main():
    print("🧠 Brain Tumor Segmentation - Complete Training System")
    print("=" * 70)
    print("📚 Multiple Models Comparison with Full Training Implementation")
    print("=" * 70)
    
    # Record start time
    start_time = time.time()
    
    # Initialize configuration and data loader
    config = Config()
    data_loader = BrainTumorDataLoader(config)
    
    # ========== REAL DATASET SETUP ==========
    # Update these paths to your actual dataset folders
    IMAGE_FOLDER = "images"      # Change this
    MASK_FOLDER = "masks"        # Change this
    
    print("Loading real brain tumor dataset...")
    data_loader.load_images_from_folders(
        image_folder=IMAGE_FOLDER,
        mask_folder=MASK_FOLDER
    )
    
    # Prepare datasets (70% train, 10% validation, 20% test)
    X_train, X_val, X_test, y_train, y_val, y_test = data_loader.prepare_datasets()
    
    print(f"\n📊 Dataset Summary:")
    print(f"   🏋️  Training set: {X_train.shape[0]} images")
    print(f"   📊 Validation set: {X_val.shape[0]} images")
    print(f"   🧪 Test set: {X_test.shape[0]} images")
    print(f"   🖼️  Image shape: {X_train.shape[1:]}")
    
    # Initialize segmenter and build all models
    segmenter = BrainTumorSegmenter(config)
    models = segmenter.build_all_models()
    
    print(f"\n🛠️  Total models built: {len(models)}")
    
    # Train all models
    print(f"\n⏰ Starting training process at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    histories = segmenter.train_models(X_train, y_train, X_val, y_val, epochs_per_model=50)
    
    # Evaluate all models
    results = segmenter.evaluate_models(X_test, y_test)
    
    # Compare and display results
    best_model_name, best_metrics = segmenter.compare_models()
    
    # Plot comprehensive comparison
    results_df = segmenter.plot_comparison()
    
    # Visualize sample predictions
    visualize_sample_predictions(segmenter, X_test, y_test, num_samples=2)
    
    # Save the best model
    best_model = segmenter.models[best_model_name]
    best_model.save('best_brain_tumor_model.h5')
    print(f"\n💾 Best model ({best_model_name}) saved as 'best_brain_tumor_model.h5'")
    
    # Calculate total execution time
    total_time = time.time() - start_time
    hours = int(total_time // 3600)
    minutes = int((total_time % 3600) // 60)
    seconds = int(total_time % 60)
    
    # Print final summary
    print(f"\n{'='*80}")
    print("🎊 TRAINING COMPLETED SUCCESSFULLY!")
    print(f"{'='*80}")
    print(f"⏱️  Total execution time: {hours:02d}:{minutes:02d}:{seconds:02d}")
    print(f"🏆 Best Model: {best_model_name}")
    print(f"📈 Global Accuracy: {best_metrics['Global_Accuracy']*100:.2f}%")
    print(f"🎯 Mean Accuracy: {best_metrics['Mean_Accuracy']*100:.2f}%")
    print(f"🔷 Mean IoU: {best_metrics['Mean_IoU']*100:.2f}%")
    print(f"📏 Mean Boundary F1: {best_metrics['Mean_Boundary_F1']*100:.2f}%")
    print(f"🎲 Dice Score: {best_metrics['Dice_Coefficient']*100:.2f}%")
    print(f"💾 Model saved: 'best_brain_tumor_model.h5'")
    print(f"{'='*80}")
    
    return segmenter, best_model_name, best_metrics, results_df

# Run the main function
if __name__ == "__main__":
    segmenter, best_model_name, best_metrics, results_df = main()
```
